```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(['clad','it','eng','cprogram','single','several','mech','phy','vdc'])
ypoints = np.array([100,60,70,50,80,40,90,30,90])

plt.plot(xpoints, ypoints)
plt.grid()
plt.show()
plt.bar(xpoints,ypoints,color='red',width=0.5)
plt.show()
plt.barh(xpoints,ypoints)
plt.show()
plt.scatter(xpoints,ypoints)
plt.show()
mylabels = ['clad','it','eng','cprogram','single','several','mech','phy','vdc']
myexplode = [0.2, 0, 0, 0.3,0,0,0,0,0]
plt.pie(ypoints,labels=mylabels,explode = myexplode)
plt.legend(title = "Semester marks")
plt.show()
x = np.random.normal(60,90,100)
plt.hist(x)
plt.show()
```
